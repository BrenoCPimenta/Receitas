```python
import pandas as pd
```

```python
data = pd.read_csv('../data/recipes_data_test.csv')
data.head(3)
```

```python
data.images = data.images.apply(lambda x: x if x[0] == '[' else str([x]))
data['preparation_time'] = data.time.apply(lambda x: int(x.split(' ')[0]))
```

```python
new_columns_names = {
    'link': 'link',
    'raw_text': 'html',
    'ingredients': 'ingredients:list',
    'text': 'raw_text',
    'page_title': 'page_title',
    'title': 'recipe_title',
    'portions': 'portions',
    'favorites': 'favorites',
    'coments': 'coments',
    'group': 'group',
    'images': 'images:list',
    'preparation_time': 'preparation_time'
}
data = data.rename(columns=new_columns_names)
data = data[list(new_columns_names.values())]
```

```python
data.to_csv('../data/recipes_data_test_index.csv', index=False)
```
